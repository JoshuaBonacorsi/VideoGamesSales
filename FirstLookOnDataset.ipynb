```python
import pandas as pd

import dash
from dash import Dash, html, dcc
import plotly.express as px
```

```python
data = pd.read_csv("data.csv")

vgsales = pd.read_csv("vgsales.csv")
vgsales.rename(columns={"Year":"Year_of_Release"},inplace=True)
```

```python
# donnees en communs
common_data = pd.merge(data,vgsales,how="inner",on=["Name","Platform","Publisher","Genre","Year_of_Release"])
common_data.head()
```

```python
common_data.columns.tolist()[:-5]
```

```python
pd.melt(common_data, 
        id_vars=common_data.columns.tolist()[:-5], 
        value_vars=common_data.columns.tolist()[-5:], 
        var_name='Localisation', 
        value_name='Ventes',
        ignore_index=False)
```

```python
common_data[["Publisher","Global_Sales"]].groupby("Publisher").sum().reset_index()
```

```python
app = Dash()
```

```python
temp = common_data[["Publisher","Global_Sales"]].groupby("Publisher").sum().reset_index()
temp.sort_values("Global_Sales",ascending=False,inplace=True)

fig = px.bar(temp.iloc[:20], 
             x="Publisher", 
             y="Global_Sales")

app.layout = html.Div(children=[
    html.H1(children='Video Games Analysis'),
    dcc.Graph(
        id='example-graph',
        figure=fig
    )
])
```

```python
app.run_server(debug=False)
```
